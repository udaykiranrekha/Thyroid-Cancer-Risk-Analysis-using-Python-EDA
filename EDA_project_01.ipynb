```python
target_col = 'diagnosis' if 'diagnosis' in df.columns else df.columns[-1]
print("\nTarget column:", target_col)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x=target_col, data=df, palette='Set2')
plt.title("Distribution of Target Variable (Thyroid Cancer Diagnosis)")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()
```

**Distribution of Numerical Features**

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

for col in numerical_cols:
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

**Gender Distribution**

```python

sns.countplot(x="Gender", data=df, palette="pastel")
plt.title("Gender Distribution")
plt.show()
```

 Family History

```python
sns.countplot(x="Family_History",data=df)
plt.title("Family History of Thyroid Cancer")
plt.show()
```

```python
sns.countplot(x="Gender", hue="Diagnosis", data=df, palette="pastel")
plt.title("Gender vs Diagnosis")
plt.show()
```

```python
sns.countplot(x="Thyroid_Cancer_Risk", data=df, order=["Low", "Medium", "High"], palette="Blues")
plt.title("Thyroid Cancer Risk Levels")
plt.show()
```

**Boxplots for Outlier Detection**

```python
for col in numerical_cols:
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

 Correlation Heatmap

```python
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

**Categorical Feature Distributions**

```python
categorical_cols = df.select_dtypes(include='object').columns.tolist()
categorical_cols = [col for col in categorical_cols if col != target_col]

for col in categorical_cols:
    sns.countplot(x=col, data=df, palette='Set3', hue=df[target_col])
    plt.title(f"Distribution of {col} by {target_col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.legend(title=target_col)
    plt.xticks(rotation=45)
    plt.show()
```
